```python
import tkinter as tk
import subprocess

def run_m1_script():
    # Run M1.py
    subprocess.run(['python', 'M1.py'])

def run_m2_script():
    # Run M2.py
    subprocess.run(['python', 'M2.py'])

window = tk.Tk()
window.title("PCB Images")
window.geometry("300x200")

run_button_m1 = tk.Button(window, text="PCB Images --- YOLO", command=run_m1_script)
run_button_m1.pack(pady=10)

run_button_m2 = tk.Button(window, text="Wafer Data --- CNN", command=run_m2_script)
run_button_m2.pack(pady=10)

window.mainloop()
```
